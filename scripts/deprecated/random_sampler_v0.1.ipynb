```python
#randomly sample n samples from each dataset
def random_sampler(input_file, output_location, n_samples): 
    #Randomly samples n samples from csv without opening file
    #Outputs file in specified location
    #Max_items will have to be guessed/found/determined beforehand

    filename = input_file
    labels = pd.read_csv(filename, sep=',', header=0, usecols=[0])
    max_items = len(labels)
    skip = sorted(random.sample(range(1, max_items),max_items-n_samples))
    df = pd.read_csv(filename, header=0, skiprows=skip)
    
    return df
```
