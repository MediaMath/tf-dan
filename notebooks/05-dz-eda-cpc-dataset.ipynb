```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob

from matplotlib import rcParams

pd.options.display.max_columns = 100

rcParams['font.family'] = 'serif'
rcParams['font.serif'] = 'times new roman'

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

# Ingest data and concatenate into single dataframe

```python
csv_paths = glob.glob('../data/raw/mm-cpc/train/data-*/*.gz')
dfs = [pd.read_csv(path) for path in csv_paths]
df = pd.concat(dfs, axis='rows')
```

```python
print(
  'The CPC dataset has {} training examples '
  'with {} features each.'.format(*df.shape)
)
```

# What is the class imbalance?

```python
class_counts = df['conversion_target'].value_counts()
class_pcts = (class_counts / class_counts.sum()) * 100
print(
  'The class balance is {:.2f}% negative '
  'and {:.2f}% positive.'.format(*class_pcts.values)
)
```

# What is the cardinality of the various features?

```python
cardinalities = df.nunique()
for index, value in cardinalities.iteritems():
  print(index.ljust(30), value)
```
